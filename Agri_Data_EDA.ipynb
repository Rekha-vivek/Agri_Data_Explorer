```python
query = """
SELECT
    a."Year",
    a."State Name",
    sum(coalesce("RICE PRODUCTION (1000 tons)",0)) AS total_rice_production
FROM agriculture_data a
where a."State Name" in(
Select 
    "State Name" from agriculture_data
GROUP BY "State Name"
ORDER BY sum(coalesce("RICE PRODUCTION (1000 tons)",0)) DESC
LIMIT 3)
GROUP BY a."Year",a."State Name"
order by a."Year";
"""

year_wise_production=pd.read_sql(query,engine)
year_wise_production.head(20)
```

```python


query_oil_2012="""
select
     "State Name",
     sum(coalesce("OILSEEDS PRODUCTION (1000 tons)",0)) as oil_2012
from agriculture_data
where "Year"=(select max("Year")-5 from agriculture_data)
group by "State Name";
"""
oilseed_rate_2012=pd.read_sql(query_oil_2012,engine)
oilseed_rate_2012
```

```python
query_oil_2017="""
select 
      "State Name",
      sum(coalesce("OILSEEDS PRODUCTION (1000 tons)",0)) as oil_2017
from agriculture_data
where "Year"=(select max("Year") from agriculture_data)
group by "State Name";
""" 
oilseed_2017=pd.read_sql(query_oil_2017,engine)
oilseed_2017
```

```python
query_oil_5_years="""
select* from(
     select 
     a."State Name",
     a.oil_2017,
     b.oil_2012,
     (a.oil_2017-b.oil_2012) as oilseed_growth
from(
   select 
      "State Name",
      sum(coalesce("OILSEEDS PRODUCTION (1000 tons)",0)) as oil_2017
from agriculture_data
where "Year"=(select max("Year") from agriculture_data)
group by "State Name") a
left join
(select
     "State Name",
     sum(coalesce("OILSEEDS PRODUCTION (1000 tons)",0)) as oil_2012
from agriculture_data
where "Year"=(select max("Year")-5 from agriculture_data)
group by "State Name") b
on a."State Name" = b."State Name") 
order by oilseed_growth desc;
""" 

oilseed_growth_rate=pd.read_sql(query_oil_5_years,engine)
oilseed_growth_rate
```

```python
plot_df = oilseed_growth_rate.sort_values(by='oilseed_growth',ascending=True)
```

```python
plt.figure(figsize=(12,10))

# Bar plot
plt.bar(
    plot_df['State Name'],
    plot_df['oilseed_growth'],
    color='cornflowerblue',
    label='Oilseed Growth'
)

# Line plot on top
plt.plot(
    plot_df['State Name'],
    plot_df['oilseed_growth'],
    color='black',
    linewidth=2,
    label='Growth Rate'
)

plt.axhline(0, color='gray', linestyle='--')  
plt.xticks(rotation=45)
plt.xlabel('State Name')
plt.ylabel('Oilseed Growth (1000 tons)')
plt.title('Oilseed Production Growth by State (Bar + Line)')
plt.legend()
plt.show()
```

```python
#4.District-wise Correlation Between Area and Production for Major Crops (Rice, Wheat, and Maize) 

rice_dist="""
select 
      "Dist Name",
      corr(
        "RICE AREA (1000 ha)",
        "RICE PRODUCTION (1000 tons)") as rice_area_production_corr
from agriculture_data
where "RICE AREA (1000 ha)" is not null and "RICE PRODUCTION (1000 tons)" is not null
group by "Dist Name";
""" 

rice_corr=pd.read_sql(rice_dist,engine)
rice_corr.head()
```

```python
wheat_dist="""
select 
      "Dist Name",
      corr(
        "WHEAT AREA (1000 ha)",
        "WHEAT PRODUCTION (1000 tons)") as wheat_area_production_corr
from agriculture_data
where "WHEAT AREA (1000 ha)" is not null and "WHEAT PRODUCTION (1000 tons)" is not null
group by "Dist Name";
""" 

wheat_corr=pd.read_sql(wheat_dist,engine)
wheat_corr.head()
```

```python
maize_dist="""
select 
      "Dist Name",
      corr(
        "MAIZE AREA (1000 ha)",
        "MAIZE PRODUCTION (1000 tons)") as maize_area_production_corr
from agriculture_data
where "MAIZE AREA (1000 ha)" is not null and "MAIZE PRODUCTION (1000 tons)" is not null
group by "Dist Name";
""" 

maize_corr=pd.read_sql(maize_dist,engine)
maize_corr.head()
```

```python
plt.figure(figsize=(20,10))
plt.hist(
    rice_corr['rice_area_production_corr'],
              bins=20,
              edgecolor='black'
)
 
plt.xlabel('Correlation between Rice Area & Production')
plt.ylabel('Number of Districts')
plt.title('Distribution of Rice Area-Production Correlation Across Districts')
plt.show()
```

```python
combined_corr=pd.concat([
    rice_corr[['rice_area_production_corr']].rename(
        columns={'rice_area_production_corr':'Correlation'}).assign(Crop='Rice'),

    wheat_corr[['wheat_area_production_corr']].rename(
        columns={'wheat_area_production_corr':'Correlation'}).assign(Crop='Wheat'),
    
    maize_corr[['maize_area_production_corr']].rename(
        columns={'maize_area_production_corr':'Correlation'}).assign(Crop='Maize')
    ],ignore_index=True)
    
plt.figure(figsize=(10,6))

sns.boxplot(
    data=combined_corr,
    x='Correlation',
    hue='Crop',
    palette='Set2'
)

plt.axhline(0, color='black', linestyle='--')
plt.xlabel('Crop')
plt.ylabel('Correlation between Area & Production')
plt.title('Area–Production Correlation by Crop (District-wise)')

plt.show()
```

```python
#10.Compare the Production of Wheat and Rice for the Top 5 States Over 10 Years

rice_and_wheat="""
select  
      "State Name"
from agriculture_data
group by "State Name"
order by sum(coalesce("RICE PRODUCTION (1000 tons)",0))+sum(coalesce("WHEAT PRODUCTION (1000 tons)",0)) desc
limit 5;
"""
rice_wheat_states=pd.read_sql(rice_and_wheat,engine)
rice_wheat_states
```

```python
year_wise_rice_wheat="""
select
      "State Name",
      "Year",
      sum(coalesce("RICE PRODUCTION (1000 tons)",0)) as rice_production,
      sum(coalesce("WHEAT PRODUCTION (1000 tons)")) as wheat_production
from agriculture_data
where "State Name" in (
        select  
      "State Name"
from agriculture_data
group by "State Name"
order by sum(coalesce("RICE PRODUCTION (1000 tons)",0))+sum(coalesce("WHEAT PRODUCTION (1000 tons)",0)) desc
limit 5)
and "Year">=(select max("Year")-10 from agriculture_data)
group by "State Name","Year" 
order by "Year","State Name";
"""
rice_wheat_year_wise=pd.read_sql(year_wise_rice_wheat,engine)
rice_wheat_year_wise.head()
```

```python
corr_data=rice_wheat_year_wise[['rice_production','wheat_production']].corr()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(corr_data,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation between Rice and Wheat Production')
plt.show()
```
